```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt("munich_temperatures_average_with_bad_data.txt")
```

```python
plt.scatter(data[:,0], data[:,1])
```

```python
clean_data = data[np.where(abs(data[:,1]) < 40)]
```

```python
plt.scatter(clean_data[:,0], clean_data[:,1])
```

```python
data_2010 = clean_data[np.where(clean_data[:,0] > 2010)]
```

```python
plt.scatter(data_2010[:,0], data_2010[:,1])
```

```python
def chepito(x, a, b, c, d):
    #return c*np.sin(b*x) + d
    return a*np.cos(b*x) + c*np.sin(b*x) + d
```

```python
time = data_2010[:,0]
Temp = data_2010[:,1]
```

```python
time
```

```python
coef, cova = curve_fit(chepito, time, Temp, p0=[1,2*np.pi,-1,10], maxfev=100000)
```

```python
coef
```

```python
x = np.linspace(2010,2013.3,100)
plt.scatter(time, Temp)
plt.plot(x,chepito(x, *coef), c='r')
#plt.plot(x,coef[0]*np.cos(coef[1]*x) + coef[2]*np.sin(coef[1]*x), c='g')
#plt.plot(x,coef[0]*np.cos(coef[1]*x), c='g')
#plt.plot(x,coef[2]*np.sin(coef[1]*x))
```
