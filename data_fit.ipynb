```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import *
from scipy.special import *
```

```python
data = np.loadtxt('some_statistical_stuff.dat')
Bins = np.array(range(21))
```

```python
y, z, w = plt.hist(data, bins=Bins, density=True)
```

```python
def Gaussiana(x, mu, sigma):
    return (1.0/np.sqrt(2*sigma**2*np.pi))*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
def Poisson(k, lamb):
    return lamb**k * np.exp(-lamb)/factorial(k)
```

```python
x = z[:-1]
params_gauss, cov_gauss = curve_fit(Gaussiana, x, y)
params_poisson, cov_poisson = curve_fit(Poisson, x, y)
```

```python
plt.hist(data, bins=Bins, density=True)
plt.plot(x, Gaussiana(x,*params_gauss), c='r')
plt.plot(x, Poisson(x,*params_poisson), c='brown')
```
